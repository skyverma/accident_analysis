```python
# Visualize the distribution of variables

# Histogram of accident count
plt.figure(figsize=(10, 6))
sns.histplot(df_full['Count'], kde=True)
plt.title('Distribution of Accident Count')
plt.xlabel('Accident Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot of cause categories
plt.figure(figsize=(10, 6))
sns.countplot(data=df_full, x="Cause category")
plt.title("Distribution of Cause Categories")
plt.xlabel("Cause Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot of outcome of incidents
plt.figure(figsize=(10,6))
sns.countplot(data=df_full, x="Outcome of Incident")
plt.title("Distribution of Outcome of Incident")
plt.xlabel("Outcome of Incident")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot of accident count by cause categories
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_full, x="Cause category", y="Count")
plt.title("Accident Count by Cause Categories")
plt.xlabel("Cause Category")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Check for missing values
print("Missing values:\n", df_full.isnull().sum())
```

```python

# Total count of accidents, injuries, and deaths
total_accidents = df_full['Count'].sum()
total_injuries = df_full.loc[df_full['Outcome of Incident'] == 'Injured', 'Count'].sum()
total_deaths = df_full.loc[df_full['Outcome of Incident'] == 'Deaths', 'Count'].sum()

print("Total accidents:", total_accidents)
print("Total injuries:", total_injuries)
print("Total deaths:", total_deaths)

# Cities with the highest number of incidents
cities_with_highest_incidents = df_full.groupby('Million Plus Cities')['Count'].sum().nlargest(5)
print("Cities with the highest number of incidents:")
print(cities_with_highest_incidents)

# Most frequent cause categories
most_frequent_cause_categories = df_full['Cause category'].value_counts().nlargest(5)
print("Most frequent cause categories:")
print(most_frequent_cause_categories)

# Most frequent cause subcategories
most_frequent_cause_subcategories = df_full['Cause Subcategory'].value_counts().nlargest(5)
print("Most frequent cause subcategories:")
print(most_frequent_cause_subcategories)
```

```python
# Heatmap: Geographic distribution of accidents
heatmap_data = df_full.groupby(['Million Plus Cities', 'Cause category'])['Count'].sum().unstack().fillna(0)
if heatmap_data.empty:
    print("Insufficient data available to generate the heatmap.")
else:
    plt.figure(figsize=(12, 10))
    sns.heatmap(heatmap_data, cmap='YlOrRd', annot=True, fmt=".0f", linewidths=0.5)
    plt.title('Geographic Distribution of Accidents by Cause Category')
    plt.xlabel('Cause Category')
    plt.ylabel('Cities')
    plt.show()
```

#Finding Rank of cities

```python
# Create pairs of (value, position) and sort by value in descending order
ranked_data = sorted(enumerate(total, start=0), key=lambda x: x[1], reverse=True)
value_lst=[]
rank_lst=[]
rankcol=[None]*51


# Print the rank list
for rank, (value, position) in enumerate(ranked_data, start=1):
    #print(f"Rank {rank}: Value {value}, Position {position}")
    value_lst.append(value)
    rank_lst.append(rank)
for i in range(50):
 rankcol[value_lst[i]]=(rank_lst[i])
```

```python
df1['Rank']=rankcol
```
